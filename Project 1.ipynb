# Filter each data according to the Service

```python
def filter_data(category):
    data = []
    
    for i in df[category]:
        if i not in data:
            data.append(i)

    for j in data:
        print("Filtered Data of Cateogory", category, f' ({j})\n\n')
        print(df[df[category] == j])
        print('\n\n\n\n')
```

```python
filter_data('ReqDay')
```
